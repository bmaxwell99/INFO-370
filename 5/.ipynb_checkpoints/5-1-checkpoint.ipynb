### 2.1

```python
a = np.array([[0, 0, 1], [0, 2, 1]])
```

```python
plot = plt.plot(a[0,:], a[1,:])
```

### 2.1.1

```python
R = lambda alpha: [[np.cos(alpha*np.pi/180), np.sin(alpha*np.pi/180)], 
                  [-np.sin(alpha*np.pi/180), np.cos(alpha*np.pi/180)]]
```

```python
A = R(45)@a
plot = plt.plot(A[0,:], A[1,:])
```

```python
A = R(-225)@a
plot = plt.plot(A[0,:], A[1,:])
```

```python
A = R(180)@a
plot = plt.plot(A[0,:], A[1,:])
```

### 2.1.2

```python
a = np.array([[6, 9, 69], 
              [4, 20, 12]])
plot = plt.plot(a[0,:], a[1,:])
plt.axes().set_aspect('equal')
```

### 2.1.3

```python
A = R(69)@a
plot = plt.plot(A[0,:], A[1,:])
plt.axes().set_aspect('equal')
```

```python
A = R(420)@a
plot = plt.plot(A[0,:], A[1,:])
plt.axes().set_aspect('equal')
```

### 2.1.4

```python
np.linalg.det(R(69))
np.linalg.det(R(420))
```

#### I assume nomatter the angle, the det will be 1 but interestingly as the angle increases past 360, a small fraction is added to the det(alpha)
