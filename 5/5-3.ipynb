### 1.1

```python
hat = pd.read_csv('crazy-hat.tsv', sep = '\t')
groups = hat.group
data = hat.drop('group', axis=1).values.T
data
```

```python
def plot(data):   
    plt.plot(figsize=(10,10))   
    plt.axes().set_aspect('equal')   
    for g in groups.unique():          
        groupData = data[:, groups == g]      
        plt.plot(groupData[0,:], groupData[1,:], c='k')    
    plt.show()
```

```python
plot(data)
```

```python
R = lambda alpha: [[np.cos(alpha*np.pi/180), np.sin(alpha*np.pi/180)], 
                  [-np.sin(alpha*np.pi/180), np.cos(alpha*np.pi/180)]]
```

```python
def flip_x(a):
    b = np.array([[-1, 0], [0, 1]])
    c = b.dot(a)
    return c

b = flip_x(data)
plot(b)
```

```python
def f(b):
    c = b.T @ a @ b
    return c
```

```python
nGrid = 30
x1 = np.linspace(-4, 4, nGrid)
x2 = np.linspace(-4, 4, nGrid)
y = np.empty((nGrid, nGrid))
for i in range(len(x1)):
    for j in range(len(x2)):
        y[i,j] = f(np.array([x1[i], x2[j] ]))
        # this is your function
fig = plt.figure()
ax = fig.add_subplot(111)
a = ax.contour(x2, x1, y)
ax.set_aspect('equal')
plt.show()
```
